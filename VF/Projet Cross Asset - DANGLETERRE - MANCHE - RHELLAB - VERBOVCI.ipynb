L'objectif de ce projet est de mettre en place des grilles de désensibilisation dans le cadre d'une gestion pilotée par horizon pour un dispositif d'épargne retraite pour des salariés via un Plan d'Epargne Retraite Collectif. L'enjeu est triple : 
- Etablir des métriques d'analyses à destination de clients qui soient compréhensibles par un grand nombre,
- Optimiser la grille équilibre de sorte à maximiser le rendement des clients tout en respectant leur choix de risque, 
- Créer une grille dynamique, à destination d'un salariat moins averse au risque et dont le PERCOL ne constitue pas l'unique source d'épargne. 

Pour répondre au problème, nous travaillerons en plusieurs étapes : 
- Tout d'abord, nous modifierons les piliers "Actions" et "Obligations" de sorte à ce que ces piliers soient diversifiés et performants, 
- Ensuite, nous établirons la présence d'une prime de diversification offerte par nos nouveaux piliers et optimiserons la grille équilibre sur la base de ce constat,
- Finalement, en tenant compte des contraintes réglementaires, nous proposerons une grille dynamique. 

```python
import pandas as pd, numpy as np 
import matplotlib.pyplot as plt
from classes.portfolio import Portfolio
from classes.visualisation import Visualisation
from classes.metrics import Metrics
from classes.pilar import Pilar
from classes.grid import Grid
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)  
warnings.filterwarnings('ignore', category=DeprecationWarning) 
```

## 1. Optimisation de la grille équilibre

L'optimisation de la grille passe d'abord par un travail sur les piliers Actions et Obligations. Ce travail vise à diversifier ces piliers et à mettre en place des stratégies permettant de récupérer du $\beta$, en essayant aussi de chercher de l'$\alpha$. 

```python
# Quelques constantes
periodicity: str = "monthly"
rebalancing: str = "quarterly"
frequency: str = "yearly"
method: str = "discret"
calculation_window:int = 6
```

### 1.1. Construction du pilier actions

Nous proposons de construire notre pilier actions en diversifiant de la manière suivante : 
- Une allocation "core" qui comprend les indices actions principaux des principales économies développées : SP500, EuroStoxx 50, Nikkei 225. Cette allocation vise principalement à récupérer le $\beta$ de marché.
- Une allocation "Momentum", basée sur la stratégie du même nom, qui contiendra des indices plus exotiques avec un double objectif : diversifier notre pilier et chercher un peu d'$\alpha$.

Chaque membre de notre pilier se verra allouer une allocation minimale, et on cherchera alors à minimiser la variance. Tout au long de cette partie, le benchmark considéré sera le MSCI Europe (MSDEE15N Index)

#### 1.1.1. Construction de la poche Momentum

Nous construisons d'abord un portefeuille Momentum sur la base des rendements mensuels cumulés, avec un rebalancement trimestriel. Le calcul des signaux est fait sur une fenêtre semestrielle. On distingue d'abord les indices core des autres, puis on créé un portefeuille Momentum sur la base des indices "non-core".

```python
core_stock_indices = ["SPX Index", "SX5E Index", "NKY Index"]
all_stock_prices = pd.read_excel("data/Inputs.xlsx", sheet_name= "Actions").set_index("Date")

# Gestion des valeurs vides par interpolation
cols_with_na = [col for col in all_stock_prices.columns if all_stock_prices[col].isnull().sum() != 0]
all_stock_prices[cols_with_na] = all_stock_prices[cols_with_na].interpolate(method = 'linear')

core_stock_prices = all_stock_prices[core_stock_indices]
not_core_prices = all_stock_prices.loc[:, ~all_stock_prices.columns.isin(core_stock_indices)]
```

```python
# Stratégie choisie : Momentum
# Méthode de sélection des actifs : classement des signaux dans l'ordre croissant et utilisation de tous les actifs.
list_strategy = ["momentum"]
list_weighting = ["ranking"]

# Création du portefeuille
momentum_stock_portfolio = Portfolio(data=not_core_prices, 
                                     periodicity = periodicity,
                                     rebalancement= rebalancing, method="discret", 
                                     list_strat=list_strategy, 
                                     list_weighting=list_weighting,
                                     calculation_window=6, asset_class="equity")
# Optimisation
momentum_stock_portfolio.run_backtest()
Visualisation.plot_weights(momentum_stock_portfolio.positions, 
                           title = "Evolution des poids, poche Momentum - Actions")
Visualisation.heatmap_correlation(momentum_stock_portfolio.returns["monthly"])
```

On affiche quelques métriques pour le portefeuille Momentum.

```python
# Récupération du benchmark, le MSDEE15N Index
stock_bench: pd.DataFrame = momentum_stock_portfolio.benchmark.iloc[calculation_window:,].iloc[:,0]

# Par souci de comparabilité, on conserve seulement les prix de 2015 à 2024.
considered_momentum_stock_ptf = momentum_stock_portfolio.portfolio_value[(momentum_stock_portfolio.portfolio_value.index >= '2015-01-01')
                                                         & (momentum_stock_portfolio.portfolio_value.index <= '2024-12-31')]
considered_bench = stock_bench[(stock_bench.index >= '2015-01-01') & (stock_bench.index <= '2024-12-31')]
stock_metrics:Metrics = Metrics(considered_momentum_stock_ptf, 
                                 method =method, frequency=frequency, benchmark=considered_bench)
print("Synthèse pour le portefeuille Momentum action : ")
premia_equity: float = stock_metrics.synthesis()
```

#### 1.1.2. Répartition core - momentum

Pour répartir notre allocation entre les différents éléments, on va : 
- imposer des poids minimaux pour chaque membre du pilier (SP500, Stoxx 50, Nikkei 225, Momentum)
- minimiser la variance.

```python
momentum_stock_portfolio.portfolio_value.name = "Momentum Ptf Value"
# Comme momentum_stock_portfolio.portfolio_value est déjà en monthly, 
# pilar_member_prices va automatiquement récupérer les prix mensuels aux bonnes dates. 
stock_pilar_member_prices = core_stock_prices.join(momentum_stock_portfolio.portfolio_value, how='inner')
```

```python
# Création du pilier
stock_pilar = Pilar(asset_prices=stock_pilar_member_prices,
                    periodicity="monthly", 
                    method="discret")

# Contraintes minimales de poids
minimal_weights_constraints = {
    "SPX Index": 0.2,
    "SX5E Index": 0.25,
    "NKY Index": 0.05,
    "Momentum Ptf Value": 0.15
}

# Programme MinVariance sur fenêtre glissante
stock_weights, stock_returns = stock_pilar.run_min_variance_optimization(
    calculation_window=calculation_window, 
    rebalancing_freq=rebalancing,
    min_weights=minimal_weights_constraints)
Visualisation.plot_weights(stock_weights)
Visualisation.heatmap_correlation(stock_pilar.returns)
```

Nous présentons quelques métriques sur le pilier final (comparé au benchmark).

```python
stock_pilar_value = stock_pilar.compute_pilar_value(stock_weights)
stock_pilar_metrics:Metrics = Metrics(
                                ptf_nav=stock_pilar_value, 
                                method =method,
                                frequency=frequency, 
                                benchmark=stock_bench)
print("Synthèse pour le pilier action : ")
premia_equity: float = stock_pilar_metrics.synthesis()
```

### 1.2. Construction du pilier obligataire

La procédure suivie est la même que pour le pilier actions. Néanmoins, nous mettons en oeuvre une stratégie combinant du Momentum et du Low Vol (50/50) pour créer notre poche factorielle.

#### 1.2.1. Construction de la poche Momentum - Low Vol

```python
# Indices core et non core
core_bond_indices = ["BTSYTRUU Index", "H16162US Index", "LEGATRUU Index"]
all_bond_prices = pd.read_excel("data/Inputs.xlsx", sheet_name= "Obligataire").set_index("Date")
# Gestion des NA par interpolation
cols_with_na = [col for col in all_bond_prices.columns if all_bond_prices[col].isnull().sum() != 0]
all_bond_prices[cols_with_na] = all_bond_prices[cols_with_na].interpolate(method = 'linear')

core_bond_prices = all_bond_prices[core_bond_indices]
not_core_bond_prices = all_bond_prices.loc[:, ~all_bond_prices.columns.isin(core_bond_indices)]

# Stratégies momentum + low vol
list_strategy = ["momentum", "low vol"]
list_weighting = ["ranking", "equalweight"]

# Calculs des poids
momentum_bond_portfolio = Portfolio(data=not_core_bond_prices,
                                    periodicity = periodicity,
                                    rebalancement= rebalancing, 
                                    method="discret", 
                                    list_strat=list_strategy, 
                                    list_weighting=list_weighting,
                                    calculation_window=6, 
                                    asset_class="bonds")
momentum_bond_portfolio.run_backtest()

# Visualisations 
Visualisation.plot_weights(momentum_bond_portfolio.positions, 
                           title = "Evolution des poids, poche Momentum + Low Vol - Obligations")
Visualisation.heatmap_correlation(momentum_bond_portfolio.returns["monthly"])

# Métriques
bond_bench: pd.DataFrame = momentum_bond_portfolio.benchmark.iloc[calculation_window:,].iloc[:,0]
metrics_bond:Metrics = Metrics(ptf_nav = momentum_bond_portfolio.portfolio_value, 
                               method =method, 
                               frequency=frequency, 
                               benchmark=bond_bench)
print("Synthèse pour le portefeuille Momentum + low vol obligations : ")
premia_bond: float = metrics_bond.synthesis()
```

#### 1.2.2. Répartition core - Momentum

A nouveau, la procédure est similaire au pilier actions. Ici, notre allocation core correspond à des indices globaux d'obligations courtes, longues, et govies.

```python
momentum_bond_portfolio.portfolio_value.name = "Momentum Low Vol Ptf Value"
# Comme momentum_bond_portfolio.portfolio_value est déjà en monthly, 
# pilar_member_prices va automatiquement récupérer les prix mensuels aux bonnes dates. 
bond_pilar_member_prices = core_bond_prices.join(momentum_bond_portfolio.portfolio_value, how='inner')

# Création du pilier
bond_pilar = Pilar(asset_prices=bond_pilar_member_prices,
                   periodicity="monthly", method="discret")
minimal_weights_constraints = {
    "BTSYTRUU Index": 0.15,
    "H16162US Index": 0.15,
    "LEGATRUU Index": 0.15, 
    "Momentum Low Vol Ptf Value": 0.15
}

# Optimisation MinVariance roulante
bond_weights, bond_returns = bond_pilar.run_min_variance_optimization(
    calculation_window=calculation_window, 
    rebalancing_freq=rebalancing,
    min_weights=minimal_weights_constraints)

Visualisation.plot_weights(bond_weights)
Visualisation.heatmap_correlation(bond_pilar.returns)
bond_pilar_value = bond_pilar.compute_pilar_value(bond_weights)
metrics_bond:Metrics = Metrics(ptf_nav = bond_pilar_value, 
                               method=method, 
                               frequency=frequency, 
                               benchmark=bond_bench)
print("Synthèse pour le pilier obligation : ")
premia_bond: float = metrics_bond.synthesis()
```

### 1.3. Pilier monétaire

```python
# Import des données
data_monetaire = pd.read_excel("data/Inputs.xlsx", sheet_name= "Monétaire").set_index("Date")
cols_with_na = [col for col in data_monetaire.columns if data_monetaire[col].isnull().sum() != 0]
data_monetaire[cols_with_na] = data_monetaire[cols_with_na].interpolate(method = 'linear')

# Sur le monétaire : pas de stratégie particulière
ptf_monetaire: pd.Series = data_monetaire

# Affichage des métriques de performance du pilier monétaire
metrics_monetaire:Metrics = Metrics(ptf_monetaire, method=method, frequency=frequency)
print("Statistiques pour le pilier monétaire : ")
metrics_monetaire.display_stats()
```

### 1.4. Optimisation de la grille sous contrainte de CVaR

Nous allons maintenant chercher à optimiser la grille initiale. Pour ce faire, nous utilisons une approche consistant à maximiser le rendement sous différentes contraintes : 
- absence de short selling, 
- la somme des poids est égale à 1, 
- **la CVaR de la nouvelle grille doit, pour chaque horizon, être au moins égale à celle de la grille initiale.**  

#### 1.4.1. Etude de la grille initiale

Nous présentons ici la grille initiale, qui nous servira de point de comparaison dans la suite. Celle-ci possède les caractéristiques attendues d'une grille de désensibilisation.

```python
initial_grid_weights = pd.read_excel("data/Inputs.xlsx", sheet_name= "Grille initiale").set_index("Horizon")

# On calcule les rendements via la méthode de Pilar
initial_assets = pd.concat([all_stock_prices["MSDEE15N Index"], all_bond_prices["LBEATREU Index"], data_monetaire["OISESTR Index"]], axis = 1).dropna()
initial_returns = Pilar(asset_prices=initial_assets, periodicity=periodicity, method = method).compute_asset_returns(periodicity)

initial_grid = Grid(
    long_term_returns=[0.06, 0.035, 0.02],
    portfolios_returns=initial_returns,
    periodicity=periodicity, 
    grid_weights=initial_grid_weights
)
Visualisation.plot_weights(initial_grid.grid_weights, 
                           title = "Evolution des poids de la grille initiale",
                           invert_xaxis= True)
Visualisation.plot_CVaR(grid = initial_grid,
                        title = "Evolution de la CVaR : grille initiale")
```

#### 1.4.2. Optimisation de la grille

La création de notre nouvelle grille se fait en 3 étapes : 
- nous récupérons nos nouveaux piliers créés précédemment
- nous calculons une prime de diversification sur la base des TRI moyens obtenus avec anciens piliers et nouveaux piliers (cf Excel)
- nous optimisons la grille avec nos hypothèses de rendement de long terme modifiées, sous contrainte que la CVaR de la grille doit être égale à celle de l'ancienne.

Tout d'abord, nous récupérons les rendements annuels de nos nouveaux piliers utilisés pour le calcul des TRI. Nous les exportons sur Excel. 

```python
metrics_stock_annual:Metrics = Metrics(stock_pilar_value, 
                               method=method, frequency="yearly", benchmark=stock_bench)
metrics_bond_annual:Metrics = Metrics(bond_pilar_value, 
                               method=method, frequency="yearly", benchmark=bond_bench)
annual_returns = pd.concat([metrics_stock_annual.returns, metrics_bond_annual.returns], axis = 1)
annual_returns.columns = ["Actions", "Obligations"]
annual_returns.to_excel("Rendements annuels des piliers.xlsx")
```

Puis, nous concaténons les rendements mensuels de nos 3 piliers pour les utiliser dans notre nouvelle grille.

```python
new_pilars_returns = pd.concat([stock_returns, bond_returns], axis = 1)
cash_returns = Metrics(ptf_monetaire, method=method, frequency="monthly").returns

# Avec l'usage de fenêtres de calcul, il y a moins de rendements des portefeuilles créés que de rendements monétaires. 
# Une harmonisation est nécessaire.
rows_to_drop = len(cash_returns) - len(new_pilars_returns)
trimmed_cash_returns = cash_returns.iloc[rows_to_drop:].copy()
trimmed_cash_returns = trimmed_cash_returns.reset_index(drop = True)
trimmed_cash_returns.index = new_pilars_returns.index
new_pilars_returns = pd.concat([new_pilars_returns, trimmed_cash_returns], axis = 1)
new_pilars_returns.columns = ["Actions", "Obligations", "Monétaire"]
```

```python
cash_annual_returns = Metrics(ptf_monetaire, method=method, frequency="yearly").returns
annual_returns = pd.concat([annual_returns, cash_annual_returns.reset_index(drop=True).iloc[1:,:]], axis = 1)
annual_returns.index = [2015 + i for i in range(11)]
Visualisation.plot_cumulative_returns(annual_returns, 
                                      title="Rendements annuels cumulés des nouveaux piliers")
```

Nous pouvons finalement lancer l'optimisation. Les hypothèses de rendement de long terme ont été augmentées de la prime de diversification calculée sur le fichier Excel associée à ce code. Nous faisons l'hypothèse, assez forte, que cette prime est équirépartie entre nos piliers actions et obligations. 

```python
new_grid = Grid([0.06563, 0.04063, 0.02], new_pilars_returns, "monthly", initial_grid.grid_weights)
new_grid = new_grid.optimize_grid(initial_grid)
Visualisation.plot_weights(new_grid.grid_weights, invert_xaxis=True)
Visualisation.plot_CVaR(new_grid, compare_with=initial_grid)
```
